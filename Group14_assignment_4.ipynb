Import libraries


```python
import cv2
import numpy as np
```

Task 1: Load the Image

```python
print("--- Task 1: Loading Image ---")
image = cv2.imread('Assignment_Image.jpg')

if image is None:
    print("Error: Could not load image. Please check that 'Assignment_Image.jpeg' is in the correct folder.")
else:
    print("Image loaded successfully!")
```

Task 2: Convert to Grayscale

```python
print("--- Task 2: Grayscale Conversion ---")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('task2_grayscale.jpg', gray_image)
```

Task 3: Resize Maintaining Aspect Ratio

```python
print("--- Task 3: Resizing ---")
scale_percent = 50
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dimensions = (width, height)

resized_image = cv2.resize(image, dimensions, interpolation=cv2.INTER_AREA)
cv2.imwrite('task3_resized.jpg', resized_image)
```

Task 4: Gaussian Blur with Different Kernels

```python
print("--- Task 4: Gaussian Blurring ---")
blur_3x3 = cv2.GaussianBlur(image, (3, 3), 0)
blur_5x5 = cv2.GaussianBlur(image, (5, 5), 0)
blur_7x7 = cv2.GaussianBlur(image, (7, 7), 0)

cv2.imwrite('task4_blur_3x3.jpg', blur_3x3)
cv2.imwrite('task4_blur_5x5.jpg', blur_5x5)
cv2.imwrite('task4_blur_7x7.jpg', blur_7x7)
```

Task 5: Canny Edge Detection

```python
print("--- Task 5: Edge Detection ---")
edges_low = cv2.Canny(gray_image, 30, 100)
edges_high = cv2.Canny(gray_image, 100, 200)

cv2.imwrite('task5_edges_low_thresh.jpg', edges_low)
cv2.imwrite('task5_edges_high_thresh.jpg', edges_high)
```

Task 6: Draw Shapes (Rectangles and Circles)

```python
print("--- Task 6: Drawing Shapes ---")
shapes_image = image.copy()
cv2.rectangle(shapes_image, (50, 50), (200, 150), (0, 0, 255), 2)
cv2.rectangle(shapes_image, (250, 50), (400, 150), (0, 0, 255), 5)
cv2.rectangle(shapes_image, (450, 50), (600, 150), (0, 0, 255), -1)

cv2.circle(shapes_image, (125, 300), 30, (0, 255, 0), 2)
cv2.circle(shapes_image, (325, 300), 60, (0, 255, 0), 4)
cv2.circle(shapes_image, (525, 300), 90, (0, 255, 0), -1)

cv2.imwrite('task6_shapes.jpg', shapes_image)
```

Task 7: Gray Level Slicing

```python
print("--- Task 7: Gray Level Slicing ---")
T1, T2 = 100, 200

slicing_no_bg = np.zeros_like(gray_image)
mask = (gray_image >= T1) & (gray_image <= T2)
slicing_no_bg[mask] = 255
cv2.imwrite('task7_slicing_no_bg.jpg', slicing_no_bg)

slicing_bg = gray_image.copy()
slicing_bg[mask] = 255
cv2.imwrite('task7_slicing_with_bg.jpg', slicing_bg)
```
